<h1>Introduction: Bank Marketing Campaign Analysis</h1>

This project is focuses on exploring and preparing data collected from direct marketing campaigns conducted by a Portuguese banking institution. These campaigns were primarily carried out through phone calls with the main objective of convincing clients to subscribe to a term deposit (the target variable in the dataset).

The dataset often referred to as the Bank Marketing Dataset that contains various details about clients, including their personal information (such as age, job and education), financial status (like account balance, loan status and housing information) and campaign-related attributes (such as contact type, number of contacts and previous outcomes).

By analyzing these attributes, the goal of this project is to understand how different factors influence a customer’s decision to subscribe to a term deposit.

<h2>Objectives of the Project</h2>

1.Load and understand the dataset to get a clear idea of its size, structure, data types, etc.

2.Clean and prepare the data by fixing missing values, duplicates and incorrect formats.

3.Analyze the data to find important patterns and relationships.

4.Generate insights that help the bank run more effective marketing campaigns.

<h2>Importance of the Study</h2>

This study helps banks understand which types of customers are more likely to open a term deposit. By knowing this, banks can focus their marketing on the right people, save money on unwanted calls and improve customer satisfaction with more personalized offers.

<h1>Dataset Overview</h1>

The dataset used in this project is the Bank Marketing Dataset, originally collected from a Portuguese banking institution and publicly available on the UCI Machine Learning Repository.

Data Source:https://archive.ics.uci.edu/dataset/222/bank+marketing

It includes approximately 45000+ records across 17 attributes, covering customer demographics, account balances, contact details and campaign results.
The data is clear and organized, making it useful for studying customer decisions and improving bank marketing strategies.

| Feature        | Description                                                            |

<h1>Data Cleaning & Pre-Processing</h1>

<h3>Handling Missing Values</h3>

```python
print('Count of Missing Values in Each Column')
df.isnull().sum()
```

There is no missing values in the dataset.

<h3>Replacing Unknown Values</h3>

```python
print('Replacing Unknown Values with Most Common Category (Mode)')
print('Count of Unknown Values in Each Column (Before)')
(df == 'unknown').sum()
```

```python
#Replacing unknown in job column
df['job'] = df['job'].replace('unknown', df['job'].mode()[0])
#Replacing unknown in education column
df['education'] = df['education'].replace('unknown', df['education'].mode()[0])
#Replacing unknown in contact column
df['contact'] = df['contact'].replace('unknown', df['contact'].mode()[0])
#Replacing unknown in poutcome column
df['poutcome'] = df['poutcome'].replace('unknown', df['poutcome'].mode()[0])
```

```python
print('Count of Unknown Values in Each Column (After)')
(df == 'unknown').sum()
```

After cleaning, all columns now have no unknown values, except `poutcome`. The unknowns in `poutcome` are the real values recorded in the dataset.

<h3>Removing Duplicates</h3>

```python
print("No of duplicates:", df.duplicated().sum())
```

The dataset is clean and already don't have any duplicates

<h3>Correcting Data Types</h3>

```python
print('Data Types Before Changing')
df.dtypes
```

```python
df['default'] = df['default'].astype('category')
df['housing'] = df['housing'].astype('category')
df['loan'] = df['loan'].astype('category')
df['contact'] = df['contact'].astype('category')
df['month'] = df['month'].astype('category')
df['poutcome'] = df['poutcome'].astype('category')
df['y'] = df['y'].astype('category')
```

This code changes columns like `default`, `housing`, `loan`, `contact`, `month`, `poutcome` and `y` into a category type because they have fixed options (like yes/no or success/failure). Doing this helps Python work faster, save memory and makes the data cleaner and easier to use for analysis or building models.

```python
print('Data Types After Changing')
df.dtypes
```

<h3>Renaming Column</h3>

```python
df.rename(columns={'y': 'subscription'}, inplace=True)
print(df.columns)
```

This changes the column name `y` to `subscription` to make its meaning clearer.

<h3>Creating new derived columns:</h3>

```python
def age_dive(age):
    if age <= 25:
        return '18 - 25'
    elif age <= 40:
        return '26 - 40'
    elif age <= 60:
        return '41 - 60'
    else:
        return '60 +'

df['age_group'] = df['age'].apply(age_dive)
df['age_group'] = df['age_group'].astype('category')
```

This shows the age in groups so it help to find which age category has more likely to subscribe and it also understand the customer behaviour by age category. 

```python
df['total_contacts'] = df['campaign'] + df['previous']
```

This shows the total contacts performed during the campaign and previous before the campaign. So we can see that if the total contacts are more the chance of the sucess of the campaign.

```python
df['avg_call_duration'] = (df['duration'] / df['campaign'] / 60).round(2) 
```

`avg_call_duration` calculates the average call duration in minutes for each customer and it also helps find if longer conversations usually lead to a positive response.
